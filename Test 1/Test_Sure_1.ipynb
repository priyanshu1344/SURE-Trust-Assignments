### Question 1 (NumPy) â€“ Titanic Dataset Survival Analysis

```python
# loading data
titanic_df = pd.read_csv('Titanic-Dataset.csv')
```

```python
titanic_df.head(5)
```

```python
#converting ['Age','Fare' and 'Survived'] to array format
arr = np.array(titanic_df[['Age','Fare','Survived']])
arr
```

```python
# filtering missing values
arr = arr[~np.isnan(arr).any(axis=1)]
arr
```

```python
#creating backup
arr_cleaned = arr.copy()
arr_cleaned
```

```python
# finding mean and std for age and fare
age_mean = arr[:, 0].mean()
age_std = arr[:, 0].std()

fare_mean = arr[:, 1].mean()
fare_std = arr[:, 1].std()
```

```python
# normalizing using z-score = (x- x_mean)/x_standard_deviation
arr[:, 0] = (arr[:, 0] - age_mean)/age_std
arr[:, 1] = (arr[:, 1] - fare_mean)/fare_std
```

```python
# finding survived and not survived 
not_survived = arr_cleaned[arr_cleaned[:, -1] == 0]
survived = arr_cleaned[arr_cleaned[:, -1] == 1]
```

```python
# filtering array based on survival
survived_arr = arr_cleaned[arr_cleaned[:, 2] == 1]
not_survived_arr = arr_cleaned[arr_cleaned[:, 2] == 0]
```

```python
# finding mean age and fare for survived and not survived
not_survived_mean_age = not_survived_arr[:, 0].mean()
not_survived_mean_fare = not_survived_arr[:, 1].mean()

survived_mean_age = survived_arr[:, 0].mean()
survived_mean_fare = survived_arr[:, 1].mean()
```

```python
# mean age and fare value for survived and not survived passengers
print("Not Survived - Mean Age:", not_survived_mean_age, "Mean Fare:", not_survived_mean_fare)
print("Survived     - Mean Age:", survived_mean_age, "Mean Fare:", survived_mean_fare)
```

```python
# classification where fare is low or high than mean values
fare = arr_cleaned[:, 1]
mean_fare = fare.mean()

fare_classification = np.where(fare< mean_fare, 'Low', 'High')

print(fare_classification)
```

### Question 2 (Pandas) â€“ Iris Dataset Summary Report 

```python
# loading dataset
iris_df = pd.read_excel('iris.xlsx')
iris_df.head()
```

```python
# creating column petal ratio
iris_df['petal_ratio'] = iris_df['petal.length']/iris_df['petal.width']
iris_df
```

```python
# average mean per species
iris_df.groupby('variety')['petal_ratio'].mean()
```

```python

species_high_std = iris_df.groupby('variety')['sepal.length'].std().sort_values(ascending=False).index[0]
print(f' {species_high_std} has the highest standard deviation in sepal length')
```

```python
# filtering the dataset where sepal_width is above the overall mean
iris_df[iris_df['sepal.width'] > iris_df['sepal.width'].mean()]
```

```python
# splitting dataset into two parts one with petal_ratio >= 2 and other petal_ratio < 2
df1 = iris_df[iris_df['petal.width'] < 2]
df2 = iris_df[iris_df['petal.width'] >= 2]
```

```python
#concating them vertically
pd.concat([df1, df2], axis=0)
```
